```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Each supply chain component
np.random.seed(42)
date_range = pd.date_range(start='2023-01-01', periods=12, freq='M')

# Raw material procurement info
raw_material = pd.DataFrame({'Date': date_range,
                             'Raw_Material_Cost': np.random.randint(5000, 20000, 12),
                             'Quantity_Procured': np.random.randint(200, 500, 12)})

# Production info
production = pd.DataFrame({'Date': date_range,
                           'Units_Produced': np.random.randint(1000, 3000, 12),
                           'Production_Cost': np.random.randint(10000, 50000, 12)})

# Inventory info
inventory = pd.DataFrame({'Date': date_range,
                          'Stock_Level': np.random.randint(500, 1500, 12),
                          'Holding_Cost': np.random.randint(2000, 8000, 12)})

# Transportation info
transportation = pd.DataFrame({'Date': date_range,
                               'Distance_Travelled_km': np.random.randint(1000, 5000, 12),
                               'Transportation_Cost': np.random.randint(5000, 20000, 12)})

# Distribution info
distribution = pd.DataFrame({'Date': date_range,
                             'Units_Distributed': np.random.randint(800, 2500, 12),
                             'Revenue': np.random.randint(30000, 70000, 12)})

# Please do not change this section
# Merge on the Date column
supply_chain_data = pd.merge(raw_material, production, on='Date')
supply_chain_data = pd.merge(supply_chain_data, inventory, on='Date')
supply_chain_data = pd.merge(supply_chain_data, transportation, on='Date')
supply_chain_data = pd.merge(supply_chain_data, distribution, on='Date')

#  Our analysis - new metrics
supply_chain_data['Profit'] = supply_chain_data['Revenue'] - (supply_chain_data['Raw_Material_Cost'] + 
                              supply_chain_data['Production_Cost'] + supply_chain_data['Holding_Cost'] + 
                              supply_chain_data['Transportation_Cost'])
```

```python
# Based on Month
# Trends: Raw Material Cost - Production Cost - and Revenue
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Raw_Material_Cost', data=supply_chain_data, label='Raw Material Cost')
sns.lineplot(x='Date', y='Production_Cost', data=supply_chain_data, label='Production Cost')
sns.lineplot(x='Date', y='Revenue', data=supply_chain_data, label='Revenue')
plt.title('Monthly Cost and Revenue Trends')
plt.xlabel('Date')
plt.ylabel('Cost/Revenue ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Profit Analysis
plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Profit', data=supply_chain_data, palette='coolwarm')
plt.title('Monthly Profit')
plt.xlabel('Date')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Inventory Stock Levels
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Stock_Level', data=supply_chain_data, label='Stock Level')
plt.title('Monthly Inventory Stock Levels')
plt.xlabel('Date')
plt.ylabel('Stock Level (Units)')
plt.grid(True)
plt.show()
```

```python
# Transportation Cost vs Distance Travelled
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Distance_Travelled_km', y='Transportation_Cost', data=supply_chain_data, hue='Profit', palette='viridis', size='Profit')
plt.title('Transportation Cost vs Distance Travelled')
plt.xlabel('Distance Travelled (km)')
plt.ylabel('Transportation Cost ($)')
plt.legend()
plt.grid(True)
plt.show()
```
